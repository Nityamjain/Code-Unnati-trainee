```python
data = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie', 'David'],
                   'Age': [25, 50, 35, 48],
                   'Salary': [10000, 96000, 54000, 52000]})
print(data)
```

```python
# Lets sort the DataFrame based on column names,
sorted_data = data.sort_index(axis=1, ascending=True)
print(sorted_data)
```

```python
#Sort the rows on the basis of their index
sorted_data = data.sort_index(axis=0, ascending=False)
print("\n",sorted_data)
```

```python
# Lets sort the rows of a dataframe based on Single column
sorted_data = data.sort_values(by='Age', ascending=True)
print(sorted_data)
```

## Experiment on  NBA dataset

```python
df = pd.read_csv("nba.csv")
df
```

```python
# Use iloc method to select the first five rows and all columns
df.iloc[0:4,:]
```

```python
# Use iloc method to select the first five rows and first two columns
df.iloc[1,3]
```

```python
# Let’s use the loc method to select the first five rows and first two columns
df.loc[0:4,["Name","Team"]]
```

```python
# Filtering a data frame by condition

# Lets select those row where college is Texas
df[df["College"]=='Texas']
```

```python
df[df['Position']=="PG"]
```

```python
# Filtering DataFrame using Query function

# Select those rows where age is greater than 30
df.query("Age > 30")
```

```python
# Print missing values column wise
df.isnull().sum()
```

```python

# Print total missing values in dataframe
df.isnull().sum().sum()
```

```python
# Lets  Drop rows that have at least one missing value
clean_df=df.dropna(axis=0, how="any")
clean_df
```

```python
## Drop rows that have less than 8 non-missing values
clean_df=df.dropna(axis=0, thresh=8)
clean_df
```

```python
# Drop columns that have at least one missing value
df_clean=df.dropna(axis=1)
df_clean
```

```python
# filling missing values using fillna() method
# Lets fill missing values of Salary column
df["Salary"].isnull().sum()
```

```python
# Calculate average salary  using mean() method
avg_salary=df["Salary"].mean()
avg_salary
```

```python
df["Salary"].fillna(avg_salary,inplace=True)
```

```python
# check for duplicate entries in columns Name and Age
print(df.duplicated(subset=['Name', 'Age']))
```

```python
# Lets create a group based on the college column
result = df.groupby("College")
result
```

```python

# Get a single group
result.get_group("Texas")
```

```python
# Lets calculate average age college wise
result["Age"].mean()
```

```python
# Lets sort the result
result["Age"].mean().sort_values(ascending=True)
```

```python
# Lets sort the result
result["Age"].mean().sort_values(ascending=True)
```
