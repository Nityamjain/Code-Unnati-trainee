### importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# importing inbulit data

```python
from sklearn.datasets import load_breast_cancer
 # instantiating
cancer = load_breast_cancer(as_frame=True)
# creating dataframe
df = cancer.frame

df
```

```python
### checking shape
print('Original Dataframe shape :',df.shape)
```

```python
## Input features
X = df[cancer['feature_names']]
print('Inputs Dataframe shape   :', X.shape)

```

### STANDARIZATION

```python
# Mean
X_mean = X.mean()
# Standard deviation
X_std = X.std() 
# Standardization
Z = (X - X_mean) / X_std
```

```python
Z
```

### COVARINACE MATRIX

```python
# covariance
c = Z.cov()

# Plot the covariance matrix
sns.heatmap(c)
plt.show()
```

### CALULATING THE  EIGENVALUE AND EIGNE VECTORS OF THE CONARIANCE MARTIX

```python
eigenvalues, eigenvectors = np.linalg.eig(c)
print('Eigen values:\n', eigenvalues)
print('Eigen values Shape:', eigenvalues.shape)
print('Eigen Vector Shape:', eigenvectors.shape)
```

### SORT THE EIGN VALUE AND VECTOR IN DESCENDING ORDER

```python
# Index the eigenvalues in descending order
idx = eigenvalues.argsort()[::-1]
 # Sort the eigenvalues in descending order
eigenvalues = eigenvalues[idx]
 # sort the corresponding eigenvectors accordingly
eigenvectors = eigenvectors[:,idx]
```

### CALCULATE THE CURMULATIVE EXPLAINED  VARIANCE

```python
explained_var = np.cumsum(eigenvalues) / np.sum(eigenvalues)
explained_var
```

### DETERMINING THE NUMBER OF PRINCIPAL COMPNENTS

```python
n_components = np.argmax(explained_var >= 0.50) + 1
n_components
```

### PROJECT THE DATA ON THE SELECTED PRINCIPAL COMPONENTS

```python
# PCA component or unit matrix
# PCA component or unit matrix
u = eigenvectors[:,:n_components]
pca_component = pd.DataFrame(u, index = cancer['feature_names'],columns = ['PC1','PC2'])
```

```python
# Plotting the PCA components
plt.figure(figsize=(10, 7))
sns.heatmap(pca_component, annot=True, cmap='coolwarm')
plt.title('PCA Components')
plt.show()
```
